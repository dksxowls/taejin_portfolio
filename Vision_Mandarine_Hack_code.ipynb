```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# 1. 데이터 로드 및 전처리
try:
    df = pd.read_csv("Madarine measure ment.csv")
except FileNotFoundError:
    print("오류: 파일을 찾을 수 없습니다. 파일 이름을 확인해주세요.")
    exit()

# 'BRIX' 컬럼 전처리 및 분류를 위한 범주형 목표 변수 생성
df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)

# 브릭스 평균 계산 (약 9.22%)
mean_brix = df['BRIX'].mean()

# 이진 분류를 위한 목표 변수 (Y) 생성: 평균 이상이면 1 (고당도), 아니면 0 (저당도)
Y = (df['BRIX'] >= mean_brix).astype(int).values

# 특성 변수 (X) 분리
X_df = df.drop(columns=['group', 'BRIX'])
X = X_df.values
feature_names = X_df.columns.tolist()

print(f"데이터셋 크기 (샘플 수): {len(df)}")
print(f"특성 개수: {len(feature_names)}")
print(f"브릭스 평균: {mean_brix:.2f} -> 목표 클래스 기준")

# 2. 데이터 스케일링 (SVM에서 필수)
# SVM은 거리 기반 모델이므로, 특성 스케일링이 필수적입니다.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 모델 정의 및 K-Fold 교차 검증 준비
# SVC (Support Vector Classification) 사용, kernel='linear' 설정
# C는 규제 매개변수입니다. 데이터가 적으므로 과적합 방지를 위해 적절히 설정합니다.
model = SVC(kernel='linear', C=0.1, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- SVM (선형 커널) 분류 결과 (5-Fold 교차 검증) ---")
acc_scores = []
fold_idx = 1
all_coefs = []

# 4. K-Fold 교차 검증 수행
for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    # 모델 학습
    model.fit(X_train, Y_train)

    # 예측 및 평가
    Y_pred = model.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    acc_scores.append(accuracy)

    # 계수(coef_)를 저장하여 나중에 평균 분석에 사용
    all_coefs.append(model.coef_[0])

    print(f"Fold {fold_idx} 정확도 (Accuracy): {accuracy:.4f}")
    fold_idx += 1

# 5. 결과 분석
print(f"\n=== 교차 검증 최종 결과 ===")
print(f"평균 정확도: {np.mean(acc_scores):.4f}")

# 6. 특성 관계 분석 (가중치/계수)
# K-Fold를 통해 얻은 모든 계수의 평균을 사용하여 안정적인 해석을 시도
mean_coefs = np.mean(all_coefs, axis=0)

# 특성 이름과 평균 계수를 매핑하여 출력
coefficients = pd.Series(mean_coefs, index=feature_names).sort_values(ascending=False)

print("\n--- 특성 관계 분석 (SVM 선형 가중치) ---")
print("\n** 브릭스(고당도=1) 분류에 대한 특성 영향력 (가중치) **")
print(coefficients)

# 해석:
# 1. 양의 가중치: 해당 특성이 증가할수록 '고당도(1)'로 분류될 가능성이 높아집니다.
# 2. 음의 가중치: 해당 특성이 증가할수록 '저당도(0)'로 분류될 가능성이 높아집니다.
# 3. 가중치의 절댓값이 클수록 분류 경계를 결정하는 데 중요한 특성입니다.
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# 1. 데이터 로드 및 전처리
try:
    df = pd.read_csv("Madarine measure ment.csv")
except FileNotFoundError:
    print("오류: 파일을 찾을 수 없습니다. 파일 이름을 확인해주세요.")
    exit()

df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)

# 목표 변수 (Y)와 특성 변수 (X) 분리
Y = df['BRIX'].values
X = df.drop(columns=['group', 'BRIX']).values
feature_names = df.drop(columns=['group', 'BRIX']).columns.tolist()

# 2. 모델 정의 및 K-Fold 교차 검증 준비
# n_estimators(트리 개수)와 max_depth(트리 최대 깊이)를 제한하여 과적합 방지
model_rf = RandomForestRegressor(n_estimators=50, max_depth=3, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- 랜덤 포레스트 회귀 결과 (5-Fold 교차 검증) ---")
rmse_scores_rf = []
fold_idx = 1

# 3. K-Fold 교차 검증 수행
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    model_rf.fit(X_train, Y_train)

    Y_pred = model_rf.predict(X_test)
    rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
    rmse_scores_rf.append(rmse)

    print(f"Fold {fold_idx} RMSE: {rmse:.4f}")
    fold_idx += 1

# 4. 결과 분석
print(f"\n=== 교차 검증 최종 결과 ===")
print(f"평균 RMSE: {np.mean(rmse_scores_rf):.4f}")

# 5. 특성 관계 분석 (특성 중요도)
model_rf.fit(X, Y)
print("\n--- 랜덤 포레스트 특성 중요도 분석 ---")

importance_rf = pd.Series(model_rf.feature_importances_, index=feature_names).sort_values(ascending=False)

print("\n** 브릭스(BRIX) 예측에 대한 특성 중요도 **")
print(importance_rf)

# 해석: 특성 중요도 값이 클수록 해당 특성이 모델의 예측에 미치는 영향이 큽니다.
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import os

# --- 1. 모델 학습 (Training) ---

def train_final_random_forest(file_path):
    """CSV 파일을 로드하고 규제 강화 랜덤 포레스트 모델을 학습합니다."""

    # 1.1. 데이터 로드
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"오류: 파일을 찾을 수 없습니다. 경로를 확인해주세요: {file_path}")
        return None
    except Exception as e:
        print(f"데이터 로드 중 오류 발생: {e}")
        return None

    # 1.2. 데이터 전처리 및 목표 변수 분리
    df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
    Y = df['BRIX'].values
    X_df = df.drop(columns=['group', 'BRIX'])

    # 1.3. 특성 공학 (9가지 특성 생성)
    # Total Peel Effect
    X_df['Total Peel Effect'] = (X_df['width'] + X_df['length']) - \
                                (X_df[' After removing the peel width'] + X_df['After removing the peel length'])
    # Shell Thickness Squared
    X_df['shell thickness_sq'] = X_df['shell thickness'] ** 2
    X = X_df.values

    # 1.4. 모델 정의 및 학습
    # 규제 강화된 하이퍼파라미터 적용 (Test RMSE 1.8043 달성 모델)
    model_rf = RandomForestRegressor(
        n_estimators=50,
        max_depth=2,
        min_samples_leaf=2,
        random_state=42
    )

    model_rf.fit(X, Y)

    print("\n[+] 모델 학습 완료.")
    print(f"    사용된 특성 수: {X.shape[1]}")
    print(f"    학습 데이터 샘플 수: {X.shape[0]}")

    return model_rf, X_df.columns.tolist()

# --- 2. 예측 함수 (Prediction) ---

def predict_brix(model, feature_names, input_data):
    """학습된 모델을 사용하여 사용자 입력 데이터를 예측합니다."""

    # 2.1. 사용자 입력 데이터를 DataFrame으로 변환
    # input_data: [weight, width, length, after_peel_width, after_peel_length, shell_thickness, juice_amount]

    input_df = pd.DataFrame(
        [input_data],
        columns=['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']
    )

    # 2.2. 특성 공학 (학습 시와 동일한 방식으로 9가지 특성 생성)
    input_df['Total Peel Effect'] = (input_df['width'] + input_df['length']) - \
                                    (input_df[' After removing the peel width'] + input_df['After removing the peel length'])
    input_df['shell thickness_sq'] = input_df['shell thickness'] ** 2

    # 2.3. 모델 예측
    X_predict = input_df[feature_names].values
    predicted_brix = model.predict(X_predict)[0]

    return predicted_brix

# --- 3. 실행 (Execution) ---

if __name__ == "__main__":

    # A. 모델 학습 단계
    # (주의: 파일 이름 'Madarine measure ment.csv'를 실제 파일 경로로 교체하세요.)
    file_path = "Madarine measure ment.csv"

    result = train_final_random_forest(file_path)
    if result is None:
        exit()

    final_model, final_features = result

    # B. 사용자 입력 및 예측 단계
    print("\n--- 브릭스 예측 시뮬레이션 ---")

    # 예시 1: 당도가 높을 것으로 예상되는 귤 (무겁고, 껍질이 두꺼움)
    print("\n[예시 1] 껍질이 두껍고 무게가 많이 나가는 귤 (고당도 예상)")
    # [weight, width, length, after_peel_width, after_peel_length, shell_thickness, Juice amount]
    user_input_1 = [
        110,   # weight
        63.0,  # width
        50.0,  # length
        55.0,  # After removing the peel width
        40.0,  # After removing the peel length
        2.5,   # shell thickness (두꺼움)
        40     # Juice amount
    ]

    predicted_brix_1 = predict_brix(final_model, final_features, user_input_1)
    print(f"입력 정보:\n  무게: 110g, 껍질 두께: 2.5mm")
    print(f"-> 예측된 브릭스 (BRIX): {predicted_brix_1:.2f}%")


    # 예시 2: 당도가 낮을 것으로 예상되는 귤 (가볍고, 껍질이 얇음)
    print("\n[예시 2] 껍질이 얇고 무게가 적게 나가는 귤 (저당도 예상)")
    user_input_2 = [
        50,    # weight
        49.36,  # width
        34.16,  # length
        43.83,  # After removing the peel width
        29.75,  # After removing the peel length
        2.18,   # shell thickness (얇음)
        15     # Juice amount
    ]

    predicted_brix_2 = predict_brix(final_model, final_features, user_input_2)
    print(f"입력 정보:\n  무게: 70g, 껍질 두께: 1.0mm")
    print(f"-> 예측된 브릭스 (BRIX): {predicted_brix_2:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# --- 1. 데이터 로드 및 10가지 특성 생성 (특성 공학) ---
try:
    df = pd.read_csv("Madarine measure ment.csv")
except FileNotFoundError:
    print("오류: 파일을 찾을 수 없습니다.")
    exit()

df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
Y = df['BRIX'].values
X_df = df.drop(columns=['group', 'BRIX'])

# 1.1. 기존 껍질차 특성 2가지 추가 (Total Peel Effect, shell thickness_sq)
X_df['Total Peel Effect'] = (X_df['width'] + X_df['length']) - \
                            (X_df[' After removing the peel width'] + X_df['After removing the peel length'])
X_df['shell thickness_sq'] = X_df['shell thickness'] ** 2

# 1.2. 새로운 비율 특성 1가지 추가 (Shell Ratio)
# 껍질 두께 / (전체 너비 + 전체 길이)
X_df['Shell Ratio'] = X_df['shell thickness'] / (X_df['width'] + X_df['length'])

X = X_df.values
feature_names = X_df.columns.tolist()

print(f"데이터셋 크기 (샘플 수): {len(df)}")
print(f"사용된 특성 개수: {len(feature_names)} (총 10개)")


# --- 2. 모델 정의 및 K-Fold 교차 검증 준비 ---
# 규제 강화된 하이퍼파라미터 적용 (max_depth=2, min_samples_leaf=2)
model_rf = RandomForestRegressor(
    n_estimators=50,
    max_depth=2,
    min_samples_leaf=2,
    random_state=42
)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- Shell Ratio 추가 랜덤 포레스트 결과 (5-Fold) ---")
rmse_scores_rf = []
train_rmse_scores = []
fold_idx = 1

# --- 3. K-Fold 교차 검증 수행 ---
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    model_rf.fit(X_train, Y_train)

    # Test RMSE (검증 성능)
    Y_pred_test = model_rf.predict(X_test)
    rmse_test = np.sqrt(mean_squared_error(Y_test, Y_pred_test))
    rmse_scores_rf.append(rmse_test)

    # Train RMSE (학습 성능)
    Y_pred_train = model_rf.predict(X_train)
    rmse_train = np.sqrt(mean_squared_error(Y_train, Y_pred_train))
    train_rmse_scores.append(rmse_train)

    print(f"Fold {fold_idx}: Test RMSE: {rmse_test:.4f} / Train RMSE: {rmse_train:.4f}")
    fold_idx += 1

# --- 4. 최종 결과 및 특성 중요도 분석 ---
avg_test_rmse = np.mean(rmse_scores_rf)
avg_train_rmse = np.mean(train_rmse_scores)

print(f"\n=== 교차 검증 최종 결과 ===")
print(f"평균 Test RMSE: {avg_test_rmse:.4f}")
print(f"평균 Train RMSE: {avg_train_rmse:.4f}")
print(f"이전 Test RMSE (1.8043) 대비 변화: {avg_test_rmse - 1.8043:.4f}")

model_rf.fit(X, Y)
print("\n--- 최종 랜덤 포레스트 특성 중요도 (Shell Ratio 포함) ---")

importance_rf = pd.Series(model_rf.feature_importances_, index=feature_names).sort_values(ascending=False)
print("\n** 브릭스(BRIX) 예측에 대한 최종 특성 중요도 **")
print(importance_rf)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# --- 1. 데이터 로드 및 11가지 특성 생성 (특성 공학) ---
try:
    df = pd.read_csv("Madarine measure ment.csv")
except FileNotFoundError:
    print("오류: 파일을 찾을 수 없습니다.")
    exit()

df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
Y = df['BRIX'].values
X_df = df.drop(columns=['group', 'BRIX'])

# 1.1. 기존 공학 특성 3가지 추가
X_df['Total Peel Effect'] = (X_df['width'] + X_df['length']) - \
                            (X_df[' After removing the peel width'] + X_df['After removing the peel length'])
X_df['shell thickness_sq'] = X_df['shell thickness'] ** 2
X_df['Shell Ratio'] = X_df['shell thickness'] / (X_df['width'] + X_df['length'])

# 1.2. 새로운 특성 1가지 추가: 과육 밀도 (Pulp Density)
# 주스 양이 0인 경우를 대비해 분모에 작은 값(1e-6)을 더합니다.
X_df['Pulp Density'] = X_df['weight'] / (
    X_df[' After removing the peel width'] * X_df['After removing the peel length'] * X_df['Juice amount'] + 1e-6
)

X = X_df.values
feature_names = X_df.columns.tolist()

print(f"데이터셋 크기 (샘플 수): {len(df)}")
print(f"사용된 특성 개수: {len(feature_names)} (총 11개)")


# --- 2. 모델 정의 및 K-Fold 교차 검증 준비 ---
# 규제 강화된 하이퍼파라미터 적용 (max_depth=2, min_samples_leaf=2)
model_rf = RandomForestRegressor(
    n_estimators=50,
    max_depth=2,
    min_samples_leaf=2,
    random_state=42
)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- Pulp Density 추가 랜덤 포레스트 결과 (5-Fold) ---")
rmse_scores_rf = []
train_rmse_scores = []
fold_idx = 1

# --- 3. K-Fold 교차 검증 수행 ---
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    model_rf.fit(X_train, Y_train)

    # Test RMSE (검증 성능)
    Y_pred_test = model_rf.predict(X_test)
    rmse_test = np.sqrt(mean_squared_error(Y_test, Y_pred_test))
    rmse_scores_rf.append(rmse_test)

    # Train RMSE (학습 성능)
    Y_pred_train = model_rf.predict(X_train)
    rmse_train = np.sqrt(mean_squared_error(Y_train, Y_pred_train))
    train_rmse_scores.append(rmse_train)

    print(f"Fold {fold_idx}: Test RMSE: {rmse_test:.4f} / Train RMSE: {rmse_train:.4f}")
    fold_idx += 1

# --- 4. 최종 결과 및 특성 중요도 분석 ---
avg_test_rmse = np.mean(rmse_scores_rf)
avg_train_rmse = np.mean(train_rmse_scores)

print(f"\n=== 교차 검증 최종 결과 ===")
print(f"평균 Test RMSE: {avg_test_rmse:.4f}")
print(f"평균 Train RMSE: {avg_train_rmse:.4f}")
print(f"이전 최적 Test RMSE (1.7928) 대비 변화: {avg_test_rmse - 1.7928:.4f}")

model_rf.fit(X, Y)
print("\n--- 최종 랜덤 포레스트 특성 중요도 (Pulp Density 포함) ---")

importance_rf = pd.Series(model_rf.feature_importances_, index=feature_names).sort_values(ascending=False)
print("\n** 브릭스(BRIX) 예측에 대한 최종 특성 중요도 **")
print(importance_rf)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import os

# 로그 변환을 적용할 특성 목록 (학습 및 예측 시 일관성 유지)
LOG_TRANSFORM_FEATURES = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length']

# --- 1. 특성 공학 및 로그 변환 함수 정의 ---
def engineer_and_transform_features(df):
    """모델 학습 및 예측 시 동일하게 10가지 특성을 생성하고 로그 변환을 적용합니다."""

    # 1.1. 로그 변환 적용
    # 0이 아닌 양수 값에만 로그 변환 적용
    for col in LOG_TRANSFORM_FEATURES:
        df[col] = np.log(df[col])

    # 1.2. 공학 특성 추가 (로그 변환된 특성들을 사용하여 계산)
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'])

    return df

# --- 2. 모델 학습 (Training) ---

def train_final_random_forest(file_path):
    """CSV 파일을 로드하고 최종 (로그 변환) 모델을 학습합니다."""

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
        return None, None

    # 데이터 전처리 및 목표 변수 분리
    df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
    Y = df['BRIX'].values
    X_df = df.drop(columns=['group', 'BRIX']).copy()

    # 2.1. 로그 변환 및 특성 공학 적용
    X_df = engineer_and_transform_features(X_df)

    X = X_df.values
    feature_names = X_df.columns.tolist()

    # 2.2. 모델 정의 및 학습 (최적 규제 적용: RMSE 1.7760 모델)
    model_rf = RandomForestRegressor(
        n_estimators=50,
        max_depth=2,
        min_samples_leaf=2,
        random_state=42
    )

    model_rf.fit(X, Y)

    print("\n[+] 최종 확정 모델 (로그 변환 적용) 학습 완료.")
    print(f"    최적 성능 (평균 Test RMSE): 1.7760%")

    return model_rf, feature_names

# --- 3. 예측 함수 (Prediction) ---

def predict_brix(model, feature_names, input_data):
    """학습된 모델을 사용하여 사용자 입력 데이터를 예측합니다."""

    # 3.1. 사용자 입력 데이터를 DataFrame으로 변환 (7개 원본 특성)
    # 순서: [weight, width, length, after_peel_width, after_peel_length, shell_thickness, Juice amount]
    input_df = pd.DataFrame(
        [input_data],
        columns=['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']
    )

    # 3.2. 로그 변환 및 특성 공학 적용 (학습 시와 동일한 파이프라인)
    # 반드시 원본 입력 데이터에 대해 로그 변환이 먼저 수행되어야 합니다!
    input_df = engineer_and_transform_features(input_df)

    # 3.3. 모델 예측
    X_predict = input_df[feature_names].values
    predicted_brix = model.predict(X_predict)[0]

    return predicted_brix

# --- 4. 실행 (Execution) ---

if __name__ == "__main__":

    # A. 모델 학습 단계
    file_path = "Madarine measure ment.csv"
    final_model, final_features = train_final_random_forest(file_path)

    if final_model is None:
        exit()

    # B. 사용자 입력 및 예측 단계
    print("\n--- 브릭스 예측 (최종 로그 변환 모델) 시뮬레이션 ---")

    # 사용자 입력 데이터 (실제 값이 11.3%였던 귤의 데이터 사용)
    # [weight, width, length, after_peel_width, after_peel_length, shell_thickness, Juice amount]
    user_input_data = [
        50,    # weight
        49.36,  # width
        34.16,  # length
        43.83,  # After removing the peel width
        29.75,  # After removing the peel length
        2.18,   # shell thickness (얇음)
        15     # Juice amount
    ]

    predicted_brix_final = predict_brix(final_model, final_features, user_input_data)

    print(f"\n[입력된 귤 정보]: {user_input_data}")
    print(f"-> 예측된 브릭스 (BRIX): {predicted_brix_final:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.base import BaseEstimator, RegressorMixin
import warnings

warnings.filterwarnings('ignore')

LOG_TRANSFORM_FEATURES = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length']

def engineer_and_transform_features(df):
    """최종 모델과 동일한 10가지 특성 공학 및 로그 변환 파이프라인"""

    # 로그 변환
    for col in LOG_TRANSFORM_FEATURES:
        df[col] = np.log(df[col])

    # 공학 특성
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'])

    return df

# --- Stacking을 위한 데이터 준비 ---
try:
    df = pd.read_csv("Madarine measure ment.csv")
except FileNotFoundError:
    print("오류: 파일을 찾을 수 없습니다.")
    exit()

df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
Y = df['BRIX'].values
X_df = df.drop(columns=['group', 'BRIX']).copy()
X_df = engineer_and_transform_features(X_df)
X = X_df.values
feature_names = X_df.columns.tolist()

# --- 1. Base Estimators 정의 (로그 변환 모델 기반) ---
# 최적 Random Forest 파라미터 적용
rf = RandomForestRegressor(n_estimators=50, max_depth=2, min_samples_leaf=2, random_state=42)
lasso = Lasso(alpha=0.1, random_state=42) # Lasso는 alpha 튜닝이 필요
gbr = GradientBoostingRegressor(n_estimators=100, max_depth=2, learning_rate=0.1, random_state=42)

# --- 2. Stacking Regressor 클래스 정의 (간소화된 Custom Stacking) ---
class StackingRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, estimators, meta_estimator):
        self.estimators = estimators
        self.meta_estimator = meta_estimator
        self.trained_estimators = []

    def fit(self, X, y, cv_folds=5, random_state=42):
        kf = KFold(n_splits=cv_folds, shuffle=True, random_state=random_state)
        oof_predictions = np.zeros((X.shape[0], len(self.estimators)))

        for i, estimator in enumerate(self.estimators):
            self.trained_estimators.append([]) # 각 base model의 fold별 모델 저장

            for fold_idx, (train_index, val_index) in enumerate(kf.split(X, y)):
                X_train, X_val = X[train_index], X[val_index]
                y_train = y[train_index]

                # Base model 학습
                estimator.fit(X_train, y_train)
                # Validation set 예측
                oof_predictions[val_index, i] = estimator.predict(X_val)
                self.trained_estimators[i].append(estimator)

        # Meta model 학습
        self.meta_estimator.fit(oof_predictions, y)
        return self

    def predict(self, X):
        meta_features = np.zeros((X.shape[0], len(self.estimators)))

        # 테스트 데이터에 대한 각 base model의 평균 예측값 사용
        for i, _ in enumerate(self.estimators):
            fold_predictions = np.array([model.predict(X) for model in self.trained_estimators[i]])
            meta_features[:, i] = np.mean(fold_predictions, axis=0)

        # Meta model 예측
        return self.meta_estimator.predict(meta_features)

# --- 3. Stacking 모델 학습 및 평가 ---

stacking_model = StackingRegressor(
    estimators=[rf, lasso, gbr],
    meta_estimator=LinearRegression()
)

kf_eval = KFold(n_splits=5, shuffle=True, random_state=42)
stacking_rmse_scores = []

print("\n--- Stacking Ensemble 모델 결과 (5-Fold) ---")

fold_idx = 1
for train_index, test_index in kf_eval.split(X):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    # Stacking 모델 학습 (내부적으로 K-Fold를 사용하여 OOF 생성)
    stacking_model.fit(X_train, Y_train, cv_folds=5, random_state=42)

    # Test 예측 및 RMSE 계산
    Y_pred_test = stacking_model.predict(X_test)
    rmse_test = np.sqrt(mean_squared_error(Y_test, Y_pred_test))
    stacking_rmse_scores.append(rmse_test)

    print(f"Fold {fold_idx}: Test RMSE: {rmse_test:.4f}")
    fold_idx += 1

# --- 4. 최종 결과 분석 ---
avg_stacking_rmse = np.mean(stacking_rmse_scores)

print(f"\n=== Stacking 교차 검증 최종 결과 ===")
print(f"평균 Test RMSE: {avg_stacking_rmse:.4f}")
print(f"이전 최적 Test RMSE (1.7760) 대비 변화: {avg_stacking_rmse - 1.7760:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.model_selection import KFold
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.metrics import mean_squared_error
import warnings

warnings.filterwarnings('ignore')

# 로그 변환을 적용할 특성 목록
LOG_TRANSFORM_FEATURES = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length']

# --- 1. 특성 공학 및 로그 변환 함수 ---
def engineer_and_transform_features(df, is_input=False):
    """모델 학습 및 예측 시 동일하게 10가지 특성 공학 및 로그 변환 파이프라인을 적용합니다."""

    # 로그 변환 (학습 시와 예측 시 모두 적용)
    for col in LOG_TRANSFORM_FEATURES:
        # 로그 변환은 양수 값에만 적용됩니다.
        df[col] = np.log(df[col])

    # 공학 특성 생성 (로그 변환된 특성들을 사용하여 계산)
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    # 껍질 두께는 로그 변환 대상이 아니지만, Shell Ratio 계산에는 사용됩니다.
    # 단, 분모인 width와 length는 이미 로그 변환된 값입니다.
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'])

    return df

# --- 2. Stacking Regressor 클래스 정의 (앙상블 모델) ---
class StackingRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, estimators, meta_estimator):
        self.estimators = estimators
        self.meta_estimator = meta_estimator
        self.trained_estimators = [] # 각 Fold에서 학습된 Base Model을 저장

    def fit(self, X, y, cv_folds=5, random_state=42):
        # 1단계: Base Model 학습 및 Out-Of-Fold (OOF) 예측 생성
        kf = KFold(n_splits=cv_folds, shuffle=True, random_state=random_state)
        oof_predictions = np.zeros((X.shape[0], len(self.estimators)))
        self.trained_estimators = [] # 초기화

        for i, estimator in enumerate(self.estimators):
            self.trained_estimators.append([])
            for fold_idx, (train_index, val_index) in enumerate(kf.split(X, y)):
                X_train, X_val = X[train_index], X[val_index]
                y_train = y[train_index]

                # Base model 학습
                estimator.fit(X_train, y_train)
                # Validation set 예측 (Meta Model의 특성으로 사용)
                oof_predictions[val_index, i] = estimator.predict(X_val)
                self.trained_estimators[i].append(estimator)

        # 2단계: Meta Model 학습 (전체 데이터를 사용하여 최종 학습)
        self.meta_estimator.fit(oof_predictions, y)
        return self

    def predict(self, X):
        # 3단계: 새로운 데이터에 대한 예측
        meta_features = np.zeros((X.shape[0], len(self.estimators)))

        # 각 Base Model이 모든 Fold에서 학습된 결과를 사용하여 예측하고 평균
        for i, _ in enumerate(self.estimators):
            fold_predictions = np.array([model.predict(X) for model in self.trained_estimators[i]])
            meta_features[:, i] = np.mean(fold_predictions, axis=0) # 평균 예측값 사용

        # Meta model 예측
        return self.meta_estimator.predict(meta_features)

# --- 3. 모델 학습 및 예측 통합 실행 ---
def train_and_predict_stacking(file_path, user_input_data):

    # A. 데이터 로드 및 전처리
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
        return None

    df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
    Y = df['BRIX'].values
    X_df = df.drop(columns=['group', 'BRIX']).copy()

    # 로그 변환 및 특성 공학 적용
    X_df = engineer_and_transform_features(X_df)
    X = X_df.values
    feature_names = X_df.columns.tolist()

    # B. 모델 정의 (로그 변환 최적화 파라미터 사용)
    rf = RandomForestRegressor(n_estimators=50, max_depth=2, min_samples_leaf=2, random_state=42)
    lasso = Lasso(alpha=0.1, random_state=42)
    gbr = GradientBoostingRegressor(n_estimators=100, max_depth=2, learning_rate=0.1, random_state=42)

    stacking_model = StackingRegressor(
        estimators=[rf, lasso, gbr],
        meta_estimator=LinearRegression()
    )

    # C. Stacking 모델 최종 학습 (전체 데이터 사용)
    # StackingRegresor 내부의 fit 함수는 전체 X에 대해 OOF를 생성하므로,
    # 최종 학습을 위해 X와 Y 전체를 사용합니다. (cv_folds=5)
    stacking_model.fit(X, Y, cv_folds=5, random_state=42)
    print("\n[+] 최종 Stacking Ensemble 모델 학습 완료.")

    # D. 사용자 예측

    # 1. 사용자 입력 데이터를 DataFrame으로 변환 (7개 원본 특성)
    input_df = pd.DataFrame(
        [user_input_data],
        columns=['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']
    )

    # 2. 로그 변환 및 특성 공학 적용 (학습 데이터와 동일하게 처리)
    input_df_processed = engineer_and_transform_features(input_df.copy())

    # 3. 모델 예측
    # feature_names 순서에 맞춰 예측 특성 데이터 추출
    X_predict = input_df_processed[feature_names].values
    predicted_brix = stacking_model.predict(X_predict)[0]

    return predicted_brix

# --- 4. 실행 (Execution) ---

if __name__ == "__main__":

    file_path = "Madarine measure ment.csv"

    # 사용자 입력 데이터 (예: 실제 값이 11.3%였던 귤의 데이터)
    # [weight, width, length, after_peel_width, after_peel_length, shell_thickness, Juice amount]
    user_input_example = [
        50,    # weight
        49.36,  # width
        34.16,  # length
        43.83,  # After removing the peel width
        29.75,  # After removing the peel length
        2.18,   # shell thickness (얇음)
        15     # Juice amount
    ]

    print("--- Stacking Ensemble 모델을 이용한 브릭스 예측 ---")
    print(f"[입력된 귤 정보]: {user_input_example}")

    final_prediction = train_and_predict_stacking(file_path, user_input_example)

    if final_prediction is not None:
        print(f"-> 예측된 브릭스 (BRIX): {final_prediction:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import os

# --- 1. 특성 공학 함수 정의 (총 13개 특성) ---
def engineer_features(df):
    """모델 학습 및 예측 시 동일하게 13가지 특성을 생성합니다."""

    # 1.1. 기존 공학 특성 (3개)
    # Total Peel Effect: 전체 크기 대비 과육 크기의 차이
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                             (df[' After removing the peel width'] + df['After removing the peel length'])
    # Shell Thickness Squared: 껍질 두께의 비선형 효과
    df['shell thickness_sq'] = df['shell thickness'] ** 2

    # Shell Ratio: 껍질 두께가 전체 크기에서 차지하는 비율
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'] + 1e-6) # 0으로 나누기 방지

    # 1.2. 새로운 공학 특성 (3개 추가)

    # Density (밀도): 무게를 부피(너비*길이로 근사)로 나눔
    # 0으로 나누는 것을 방지하기 위해 분모에 작은 값(1e-6)을 더합니다.
    df['Density'] = df['weight'] / (df['width'] * df['length'] + 1e-6)

    # Aspect_Ratio (종횡비): 과일의 모양 (길이 / 너비)
    df['Aspect_Ratio'] = df['length'] / (df['width'] + 1e-6)

    # Juice_Yield_Weight (무게 대비 주스 수율)
    df['Juice_Yield_Weight'] = df['Juice amount'] / (df['weight'] + 1e-6)

    return df

# --- 2. 모델 학습 (Training) ---

def train_final_random_forest(file_path):
    """CSV 파일을 로드하고 13가지 특성 기반의 최종 모델을 학습합니다."""

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
        return None, None

    # 데이터 전처리 및 목표 변수 분리
    df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
    Y = df['BRIX'].values
    X_df = df.drop(columns=['group', 'BRIX']).copy()

    # 2.1. 특성 공학 적용 (총 13개 특성 생성)
    X_df = engineer_features(X_df)

    X = X_df.values
    feature_names = X_df.columns.tolist()

    # 2.2. 모델 정의 및 학습 (최적 규제 적용: RMSE 1.7928 모델)
    # 참고: 특성이 10개 -> 13개로 늘어났으므로,
    #       최적의 하이퍼파라미터(max_depth 등)가 달라질 수 있습니다.
    #       여기서는 비교를 위해 기존 파라미터를 유지합니다.
    model_rf = RandomForestRegressor(
        n_estimators=50,
        max_depth=2,
        min_samples_leaf=2,
        random_state=42
    )

    model_rf.fit(X, Y)

    print(f"\n[+] 최종 {len(feature_names)}가지 특성 모델 학습 완료.")
    # print(f"    최적 성능 (평균 Test RMSE): 1.7928%") # 이 값은 10개 기준이므로 주석 처리
    print(f"    사용된 특성 ({len(feature_names)}개): {feature_names}")

    return model_rf, feature_names

# --- 3. 예측 함수 (Prediction) ---

def predict_brix(model, feature_names, input_data):
    """학습된 모델을 사용하여 사용자 입력 데이터를 예측합니다."""

    # 3.1. 사용자 입력 데이터를 DataFrame으로 변환 (7개 원본 특성)
    # 순서: [weight, width, length, after_peel_width, after_peel_length, shell_thickness, Juice amount]
    input_df = pd.DataFrame(
        [input_data],
        columns=['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']
    )

    # 3.2. 특성 공학 적용 (학습 시와 동일한 6가지 특성 추가 -> 총 13개)
    input_df = engineer_features(input_df)

    # 3.3. 모델 예측
    # 반드시 feature_names의 순서대로 특성 데이터를 정렬해야 합니다.
    X_predict = input_df[feature_names].values
    predicted_brix = model.predict(X_predict)[0]

    return predicted_brix

# --- 4. 실행 (Execution) ---

if __name__ == "__main__":

    # A. 모델 학습 단계
    file_path = "Madarine measure ment.csv"
    final_model, final_features = train_final_random_forest(file_path)

    if final_model is None:
        exit()

    # B. 사용자 입력 및 예측 단계
    print(f"\n--- 브릭스 예측 (최종 {len(final_features)}가지 특성 모델) 시뮬레이션 ---")

    # 사용자 입력 데이터 (실제 값이 11.3%였던 귤의 데이터 사용)
    # [weight, width, length, after_peel_width, after_peel_length, shell_thickness, Juice amount]
    user_input_data = [
        50,    # weight
        49.36,  # width
        34.16,  # length
        43.83,  # After removing the peel width
        29.75,  # After removing the peel length
        2.18,   # shell thickness (얇음)
        15     # Juice amount
    ]

    predicted_brix_final = predict_brix(final_model, final_features, user_input_data)

    print(f"\n[입력된 귤 정보]: {user_input_data}")
    print(f"-> 예측된 브릭스 (BRIX): {predicted_brix_final:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import os

# --- 1. 특성 공학 함수 정의 (총 13개 특성) ---
def engineer_features(df):
    """모델 학습 및 예측 시 동일하게 13가지 특성을 생성합니다."""

    # 1.1. 기존 공학 특성 (3개)
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                             (df[' After removing the peel width'] + df['After removing the peel length'])
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'] + 1e-6) # 0으로 나누기 방지

    # 1.2. 새로운 공학 특성 (3개 추가)
    df['Density'] = df['weight'] / (df['width'] * df['length'] + 1e-6)
    df['Aspect_Ratio'] = df['length'] / (df['width'] + 1e-6)
    df['Juice_Yield_Weight'] = df['Juice amount'] / (df['weight'] + 1e-6)

    return df

# --- 2. 모델 학습 (Training) with GridSearchCV ---

def train_final_random_forest(file_path):
    """
    CSV 파일을 로드하고 13가지 특성 기반의 모델을 학습합니다.
    GridSearchCV를 사용하여 최적의 하이퍼파라미터를 탐색합니다.
    """

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
        return None, None

    # 데이터 전처리 및 목표 변수 분리
    df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
    Y = df['BRIX'].values
    X_df = df.drop(columns=['group', 'BRIX']).copy()

    # 2.1. 특성 공학 적용 (총 13개 특성 생성)
    X_df = engineer_features(X_df)

    X = X_df.values
    feature_names = X_df.columns.tolist()

    # 2.2. 하이퍼파라미터 튜닝 (GridSearchCV)
    print("\n[+] 13개 특성 기반 모델의 최적 하이퍼파라미터 탐색 중...")

    # 탐색할 파라미터 그리드 정의
    # (n_estimators: 트리 개수, max_depth: 트리 깊이, min_samples_leaf: 리프 노드 최소 샘플 수)
    param_grid = {
        'n_estimators': [50, 100, 150],
        'max_depth': [2, 3, 4, 5, None], # None은 무제한 깊이
        'min_samples_leaf': [2, 3, 5]
    }

    # 기본 RandomForest 모델
    base_model = RandomForestRegressor(random_state=42)

    # GridSearchCV 설정 (cv=5: 5-Fold Cross Validation)
    # scoring='neg_mean_squared_error': RMSE와 유사한 MSE를 기준으로 평가
    grid_search = GridSearchCV(
        estimator=base_model,
        param_grid=param_grid,
        cv=5,
        scoring='neg_mean_squared_error',
        n_jobs=-1 # 사용 가능한 모든 CPU 코어 사용
    )

    # 튜닝 실행
    grid_search.fit(X, Y)

    # 2.3. 최적 모델 및 결과 확인
    final_model = grid_search.best_estimator_
    best_params = grid_search.best_params_

    # GridSearchCV는 음수 MSE를 반환하므로, -1을 곱하고 제곱근(sqrt)을 취해 RMSE 계산
    best_rmse = np.sqrt(-grid_search.best_score_)

    print("\n[+] 최적 하이퍼파라미터 탐색 완료.")
    print(f"    최적의 파라미터: {best_params}")
    print(f"    최적 성능 (평균 CV RMSE): {best_rmse:.4f}%")
    print(f"    사용된 특성 ({len(feature_names)}개): {feature_names}")

    return final_model, feature_names

# --- 3. 예측 함수 (Prediction) ---
# (이전 코드와 동일 - 수정 없음)
def predict_brix(model, feature_names, input_data):
    """학습된 모델을 사용하여 사용자 입력 데이터를 예측합니다."""

    input_df = pd.DataFrame(
        [input_data],
        columns=['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']
    )

    input_df = engineer_features(input_df)

    X_predict = input_df[feature_names].values
    predicted_brix = model.predict(X_predict)[0]

    return predicted_brix

# --- 4. 실행 (Execution) ---

if __name__ == "__main__":

    # A. 모델 학습 단계 (GridSearchCV 실행)
    file_path = "Madarine measure ment.csv"
    final_model, final_features = train_final_random_forest(file_path)

    if final_model is None:
        exit()

    # B. 사용자 입력 및 예측 단계
    print(f"\n--- 브릭스 예측 (최종 {len(final_features)}가지 특성 + 튜닝 모델) 시뮬레이션 ---")

    user_input_data = [
        105,   # weight (g)
        62.49, # width (mm)
        50.53, # length (mm)
        53.87, # After removing the peel width (mm)
        37.95, # After removing the peel length (mm)
        2.42,  # shell thickness (mm)
        43     # Juice amount (ml)
    ]

    predicted_brix_final = predict_brix(final_model, final_features, user_input_data)

    print(f"\n[입력된 귤 정보]: {user_input_data}")
    print(f"-> 예측된 브릭스 (BRIX): {predicted_brix_final:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import warnings

# 경고 무시 설정
warnings.filterwarnings('ignore')

# --- 1. 특성 공학 함수 정의 (총 10개 특성) ---
def engineer_features_10(df):
    """모델 학습 및 예측 시 동일하게 10가지 특성을 생성합니다 (로그 변환 없음)."""

    # Total Peel Effect: 전체 크기 대비 과육 크기의 차이
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    # Shell Thickness Squared: 껍질 두께의 비선형 효과
    df['shell thickness_sq'] = df['shell thickness'] ** 2

    # Shell Ratio: 껍질 두께가 전체 크기에서 차지하는 비율
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'] + 1e-6) # 0으로 나누기 방지

    return df

# --- 2. 데이터 로드 및 전처리 ---
file_path = "/content/Madarine measure ment1.csv" # Corrected file path to absolute path
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
    exit()

# 컬럼명 매핑 (요청하신 tangerine still long 포함)
df.rename(columns={
    '무게': 'weight',
    '가로': 'width',
    '세로': 'length',
    '껍질제거후 가로': ' After removing the peel width',
    '껍질제거후 세로': 'After removing the peel length',
    '껍질 두께': 'shell thickness',
    '과즙량': 'Juice amount',
    '그룹': 'group',
    '귤 꼭지 길이': 'tangerine still long' # 요청하신 컬럼명 추가
}, inplace=True)

df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
Y = df['BRIX'].values

# 'group', 'BRIX' 컬럼을 제외하고 특성 데이터프레임 생성
# 이 모델은 10가지 핵심 특성만 사용하므로 'tangerine still long'은 제외됩니다.
# Corrected column dropping: only drop 'group' and 'BRIX'
X_df = df.drop(columns=['group', 'BRIX']).copy()

# 특성 공학 적용 (10가지 특성)
X_df = engineer_features_10(X_df)
X = X_df.values
N_FEATURES = len(X_df.columns)

# --- 3. 모델 정의 및 K-Fold 교차 검증 준비 ---
# RMSE 1.7928을 달성했던 규제 강화 파라미터 사용
model_rf = RandomForestRegressor(
    n_estimators=50,
    max_depth=2,
    min_samples_leaf=2,
    random_state=42
)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 결과를 저장할 리스트 초기화
all_actual = []
all_predicted = []
all_rmse = []

# --- 4. K-Fold 교차 검증 및 결과 저장 ---
print(f"--- 10가지 특성 모델 (로그 변환 없음) 예측 결과 ---")
print("Fold별 Test RMSE 및 예측값 vs 실제값 비교")

for fold_idx, (train_index, test_index) in enumerate(kf.split(X)):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    model_rf.fit(X_train, Y_train)
    Y_pred_test = model_rf.predict(X_test)

    rmse_test = np.sqrt(mean_squared_error(Y_test, Y_pred_test))
    all_rmse.append(rmse_test)

    # 해당 Fold의 예측값과 실제값 저장
    all_actual.extend(Y_test)
    all_predicted.extend(Y_pred_test)

    print(f"\n[Fold {fold_idx + 1}] Test RMSE: {rmse_test:.4f}")

# --- 5. 최종 결과 표 출력 ---
# OOF (Out-Of-Fold) 예측 결과를 데이터프레임으로 변환
results_df = pd.DataFrame({
    'Actual_BRIX': all_actual,
    'Predicted_BRIX': all_predicted,
})

# 오차 (Error) 및 절대 오차 (Absolute Error) 계산
results_df['Error'] = results_df['Predicted_BRIX'] - results_df['Actual_BRIX']
results_df['Absolute_Error'] = np.abs(results_df['Error'])

# 오차 크기 순으로 정렬하여 출력
results_df_sorted = results_df.sort_values(by='Absolute_Error', ascending=False).reset_index(drop=True)

print("\n" + "=" * 60)
print(f"        🔍 모든 테스트 샘플의 예측값 vs 실제 값 (총 {N_FEATURES}가지 특성) 🔍")
print("=" * 60)
print(results_df_sorted.to_string(index=True, float_format="%.2f"))

print("\n" + "=" * 60)
print(f"최종 평균 Test RMSE: {np.mean(all_rmse):.4f} (로그 변환 최적 모델: 1.7614)")
print("=" * 60)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import warnings

warnings.filterwarnings('ignore')

# --- 1. 모델 파이프라인 상수 정의 ---
LOG_TRANSFORM_FEATURES = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length']

# --- 2. 특성 공학 및 로그 변환 함수 정의 (총 13개 특성) ---
def engineer_and_transform_features(df):
    """13가지 특성을 생성하고 로그 변환을 적용합니다."""

    # 1.1. 로그 변환 적용
    for col in LOG_TRANSFORM_FEATURES:
        # 0 또는 음수 값 방지 (실제 귤 데이터에서는 발생하지 않아야 함)
        df[col] = np.log(df[col])

    # 1.2. 공학 특성 (총 13개) 생성
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'] + 1e-6)

    # 새로운 공학 특성 (로그 변환된 값 사용)
    df['Density'] = df['weight'] / (df['width'] * df['length'] + 1e-6)
    df['Aspect_Ratio'] = df['length'] / (df['width'] + 1e-6)
    df['Juice_Yield_Weight'] = df['Juice amount'] / (df['weight'] + 1e-6)

    return df

# --- 3. 모델 학습 함수 ---
def train_final_model(file_path):
    """데이터를 로드, 전처리하고 최종 랜덤 포레스트 모델을 학습합니다."""

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
        return None, None # Return None for both model and feature_names

    # 컬럼명 매핑
    df.rename(columns={
        '무게': 'weight',
        '가로': 'width',
        '세로': 'length',
        '껍질제거후 가로': ' After removing the peel width',
        '껍질제거후 세로': 'After removing the peel length',
        '껍질 두께': 'shell thickness',
        '과즙량': 'Juice amount',
        '그룹': 'group',
        '귤 꼭지 길이': 'tangerine still long'
    }, inplace=True)

    # BRIX 값 전처리 (BRIX %를 비율로 변환)
    df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float)
    Y = df['BRIX'].values

    # 특성 데이터프레임 준비 및 전처리
    X_df = df.drop(columns=['group', 'BRIX', 'tangerine still long'], errors='ignore').copy()
    X_df = engineer_and_transform_features(X_df)
    X = X_df.values

    # 모델 정의 및 학습
    model_rf = RandomForestRegressor(
        n_estimators=50,
        max_depth=2,
        min_samples_leaf=2,
        random_state=42
    )
    model_rf.fit(X, Y)

    # 학습된 모델과 특성 순서를 반환
    return model_rf, X_df.columns

# --- 4. 사용자 데이터 예측 함수 ---
def predict_brix(model, feature_names, raw_data):
    """단일 측정 데이터로 BRIX를 예측합니다."""

    # Define input column names for clarity and correct DataFrame creation
    input_cols = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']

    # 입력 데이터를 DataFrame으로 변환 (컬럼명은 학습 데이터와 동일하게)
    input_df = pd.DataFrame(
        [raw_data],
        columns=input_cols # Explicitly provide column names
    )

    # 모델 학습 시 제외했던 특성 컬럼은 예측 시에도 제외 (여기서는 drop 하지 않음. 이미 input_df 생성시 제어)

    # 특성 공학 및 로그 변환 적용
    input_df = engineer_and_transform_features(input_df)

    # 학습된 특성 순서(feature_names)에 맞게 예측 데이터 정렬
    X_predict = input_df[feature_names].values

    # 예측 수행 및 결과 변환 (비율 -> %p)
    predicted_brix_ratio = model.predict(X_predict)[0]
    predicted_brix_percent = predicted_brix_ratio

    return predicted_brix_percent

# --- 5. 실행 로직 ---
file_path = "/content/Madarine measure ment1.csv"
model, feature_names = train_final_model(file_path)

if model:
    print("============================================")
    print("✅ BRIX 예측 모델 학습 완료 (RMSE ≈ 1.76%p)")
    print("============================================")

    # 📌 사용자 입력 데이터 예시
    # (실제 데이터셋의 첫 번째 행 데이터를 가져와 사용했습니다: Actual BRIX 11.2%)
    sample_data = [
        50,    # weight
        49.36,  # width
        34.16,  # length
        43.83,  # After removing the peel width
        29.75,  # After removing the peel length
        2.18,   # shell thickness (얇음)
        15     # Juice amount
    ]

    # 예측 수행
    predicted_brix = predict_brix(model, feature_names, sample_data)

    print("\n--- 귤 샘플 측정값 ---")
    # raw_data는 리스트이므로 인덱싱으로 접근
    input_cols = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length', 'shell thickness', 'Juice amount']
    for i, col in enumerate(input_cols):
        print(f"  {col}: {sample_data[i]}")

    print("\n------------------------------")
    print(f"최종 예측 BRIX 당도: {predicted_brix:.2f}%")
    print("------------------------------")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import warnings

warnings.filterwarnings('ignore')

# --- 1. 모델 파이프라인 상수 및 함수 정의 ---
LOG_TRANSFORM_FEATURES = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length']

def engineer_and_transform_features(df):
    """13가지 특성을 생성하고 로그 변환을 적용합니다."""

    for col in LOG_TRANSFORM_FEATURES:
        df[col] = np.log(df[col])

    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'] + 1e-6)

    df['Density'] = df['weight'] / (df['width'] * df['length'] + 1e-6)
    df['Aspect_Ratio'] = df['length'] / (df['width'] + 1e-6)
    df['Juice_Yield_Weight'] = df['Juice amount'] / (df['weight'] + 1e-6)

    return df

# --- 2. 데이터 로드 및 전처리 ---
file_path = "/content/Madarine measure ment1.csv"
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
    exit()

df.rename(columns={
    '무게': 'weight', '가로': 'width', '세로': 'length',
    '껍질제거후 가로': ' After removing the peel width', '껍질제거후 세로': 'After removing the peel length',
    '껍질 두께': 'shell thickness', '과즙량': 'Juice amount',
    '그룹': 'group', '귤 꼭지 길이': 'tangerine still long'
}, inplace=True)

# BRIX 값을 float으로 변환 (비율 형태, 0.xx)
df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float) / 100
Y = df['BRIX'].values

# 특성 데이터프레임 준비 및 전처리
X_df = df.drop(columns=['group', 'BRIX', 'tangerine still long']).copy()
X_df = engineer_and_transform_features(X_df)
X = X_df.values

# --- 3. K-Fold 교차 검증 및 예측값 저장 ---
# Stacking의 Level 1 모델 학습을 위한 Out-of-Fold 예측값을 생성합니다.
model_rf = RandomForestRegressor(n_estimators=50, max_depth=2, min_samples_leaf=2, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Level 1 모델의 입력으로 사용할 OOF 예측값 배열 초기화
oof_predictions = np.zeros(len(X))
oof_rmse = []

print("--- [Level 0] Random Forest (물리적 특성) OOF 예측값 생성 중 ---")

for fold_idx, (train_index, test_index) in enumerate(kf.split(X)):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    model_rf.fit(X_train, Y_train)
    Y_pred_test = model_rf.predict(X_test)

    # 해당 폴드의 예측값을 전체 예측 배열의 해당 인덱스에 저장 (Stacking의 핵심)
    oof_predictions[test_index] = Y_pred_test

    rmse_test = np.sqrt(mean_squared_error(Y_test, Y_pred_test))
    oof_rmse.append(rmse_test * 100)
    print(f"[Fold {fold_idx + 1}] Test RMSE: {rmse_test * 100:.2f}%p")

# --- 4. 결과 DataFrame 생성 및 저장 ---
df_stacking_input = pd.DataFrame({
    'Actual_BRIX': Y,
    'RF_Pred': oof_predictions
})

# Stacking을 위해 RF 예측값을 CSV 파일로 저장
csv_file_path = 'RF_Stacking_Input_PA.csv'
df_stacking_input.to_csv(csv_file_path, index=False)

print("\n" + "=" * 60)
print(f"최종 RF 평균 OOF RMSE: {np.mean(oof_rmse):.2f}%p")
print(f"✅ Stacking을 위한 RF 예측값 파일이 '{csv_file_path}'에 저장되었습니다.")
print("=" * 60)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import warnings

warnings.filterwarnings('ignore')

# --- 1. 모델 파이프라인 상수 및 함수 정의 ---
LOG_TRANSFORM_FEATURES = ['weight', 'width', 'length', ' After removing the peel width', 'After removing the peel length']
INPUT_FEATURE_KEYS = [
    'weight', 'width', 'length', ' After removing the peel width',
    'After removing the peel length', 'shell thickness', 'Juice amount'
]

def engineer_and_transform_features(df):
    """5가지 로그 변환 및 3가지 공학 특성만 생성합니다 (총 10개)."""

    # 1.1. 로그 변환 적용 (5개)
    for col in LOG_TRANSFORM_FEATURES:
        df[col] = np.log(df[col])

    # 1.2. 공학 특성 (3개) 생성
    df['Total Peel Effect'] = (df['width'] + df['length']) - \
                                (df[' After removing the peel width'] + df['After removing the peel length'])
    df['Total width'] = df['width'] - df[' After removing the peel width']
    df['Total length'] = df['length'] - df['After removing the peel length']
    df['shell thickness_sq'] = df['shell thickness'] ** 2
    df['Shell Ratio'] = df['shell thickness'] / (df['width'] + df['length'] + 1e-6)

    return df

# --- 2. 데이터 로드 및 전처리 ---
file_path = "/content/Madarine measure ment1.csv"
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"오류: 파일을 찾을 수 없습니다: {file_path}")
    exit()

# 컬럼명 매핑 및 BRIX 변환
df.rename(columns={
    '무게': 'weight', '가로': 'width', '세로': 'length',
    '껍질제거후 가로': ' After removing the peel width', '껍질제거후 세로': 'After removing the peel length',
    '껍질 두께': 'shell thickness', '과즙량': 'Juice amount',
    '그룹': 'group', '귤 꼭지 길이': 'tangerine still long'
}, inplace=True)

df['BRIX'] = df['BRIX'].astype(str).str.replace('%', '', regex=False).astype(float) / 100
Y = df['BRIX'].values
N_SAMPLES = len(Y)

X_df = df.drop(columns=['group', 'BRIX', 'tangerine still long','Juice amount']).copy()
X_df = engineer_and_transform_features(X_df)
X = X_df.values

# --- 3. 모델 정의 및 K-Fold 교차 검증 실행 ---
model_rf_cv = RandomForestRegressor(
    n_estimators=50,
    max_depth=2,
    min_samples_leaf=2,
    random_state=42
)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

fold_rmse_list = []
# OOF (Out-of-Fold) 예측값을 저장하는 배열
oof_predictions = np.zeros(N_SAMPLES)
print("--- 🍊 K-Fold 교차 검증 (일반화 성능 평가) ---")

for fold_idx, (train_index, test_index) in enumerate(kf.split(X)):
    model_rf_cv.fit(X[train_index], Y[train_index])
    Y_pred_test = model_rf_cv.predict(X[test_index])

    # OOF 예측값 저장 (테스트 데이터의 예측값)
    oof_predictions[test_index] = Y_pred_test

    rmse_test = np.sqrt(mean_squared_error(Y[test_index], Y_pred_test))
    fold_rmse_list.append(rmse_test * 100)
    print(f"[Fold {fold_idx + 1}] Test RMSE: {rmse_test * 100:.2f}%p")

avg_rmse = np.mean(fold_rmse_list)
print("\n" + "=" * 60)
print(f"✅ 평균 K-Fold Test RMSE: {avg_rmse:.2f}%p")
print("=" * 60)

# --- 4. 테스트 데이터의 예측값, 실제값 및 오차 출력 ---
# 오차(Error) = 실제값 - 예측값
results_df = pd.DataFrame({
    'Sample_ID': df.index.values, # 데이터 프레임 인덱스를 샘플 ID로 사용
    'Actual_BRIX (%)': Y * 100,
    'Predicted_BRIX (%)': oof_predictions * 100
})
results_df['Error (%)'] = results_df['Actual_BRIX (%)'] - results_df['Predicted_BRIX (%)']


print("\n--- 📊 전체 OOF (Test Set) 예측값, 실제값, 오차 ---")
print(results_df.round(4).to_markdown(index=False)) # 모든 샘플 출력
print("-" * 60)
print(f"전체 Test Set RMSE (OOF 기준): {np.sqrt(mean_squared_error(Y, oof_predictions)) * 100:.4f}%p")


# --- 5. 최종 모델 학습 (전체 데이터 사용) 및 사용자 데이터 예측 ---
model_rf_final = RandomForestRegressor(
    n_estimators=50,
    max_depth=2,
    min_samples_leaf=2,
    random_state=42
)
model_rf_final.fit(X, Y)
print("\n✅ 최종 예측 모델 학습 완료 (전체 데이터 사용)")

# 📌 사용자 입력 데이터
user_input_data = [
    50, 49.36, 34.16, 43.83, 29.75, 2.18, 15
]

raw_data_dict = dict(zip(INPUT_FEATURE_KEYS, user_input_data))
input_df = pd.DataFrame([raw_data_dict])

X_predict_df = engineer_and_transform_features(input_df.copy())
X_predict = X_predict_df[X_df.columns].values

Yp_ratio = model_rf_final.predict(X_predict)[0]
Yp_percent = Yp_ratio * 100

print("\n" + "=" * 60)
print("              🌟 사용자 데이터 예측 결과 🌟")
print("-" * 60)
print(f"입력 데이터: {user_input_data}")
print(f"최종 예측 BRIX 당도: {Yp_percent:.4f}%")
print("=" * 60)
```
